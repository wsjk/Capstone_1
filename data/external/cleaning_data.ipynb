```python
import pandas as pd
import numpy as np
import glob
import os
import csv
import re
from pathlib import PureWindowsPath
```

```python
cwd = PureWindowsPath(os.getcwd())
raw_data_path = os.path.join(cwd.parents[0],'raw')
raw_data_path
```

```python
data_files = glob.glob(os.path.join(raw_data_path, '*.csv'), recursive=False)
data_files
```

```python
imdb_df = pd.read_csv(data_files[1], header=0)
imdb_df.head()
```

```python
def check_csv(csvfile):
    bad_lines = {}
    for row, line in enumerate(csvfile):
        if len(line) > 44:
            bad_lines[row] = ','.join(line)
    return bad_lines

csvfile = []
with open(data_files[0], mode='r',encoding='utf8') as csvfile:
    reader = csv.reader(csvfile)
    csvfile = list(reader)

bad_lines = check_csv(csvfile)
print(len(bad_lines.keys()))
```

```python
for k,v in bad_lines.items():
    bad_lines[k] = v.replace("\\","")

new_csvfile = csvfile
for row, line in enumerate(new_csvfile):
    if row in bad_lines.keys():
        new_csvfile[row] = bad_lines[row].split(',')

bad_lines = check_csv(new_csvfile)
print(len(bad_lines.keys()))

```

```python
#indexes = [8127,9388]
#for index in sorted(indexes, reverse=True):
#    del csvfile[index]

def fix_titles(pattern, bad_lines):
    del_lines = []
    for k,v in bad_lines.items():
        pattern_match = pattern.findall(v) 
        
        #del lines with unacceptable titles
        if not pattern_match:
            del_lines.append(k)
            continue
            
        title = pattern_match[0][3]
        new_title = re.sub(',','', title)
        v = re.sub(title, new_title, v)
        bad_lines[k]=v
    
    return del_lines, bad_lines
    
pattern = re.compile( r'(titles\d+)\/(.+),(tt\d+),(.+) \((.+)\)(.+)')

del_lines, bad_lines = fix_titles(pattern, bad_lines)

print(del_lines)
print(bad_lines)
```

```python
with open(data_files[1],'w') as clean_file:
    writer = csv.writer(clean_file)
    writer.writerows(new_lines)
```
