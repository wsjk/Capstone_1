# H2O Random Forests

```python
import h2o
```

```python
h2o.init()
```

```python
df = h2o.import_file('mega_features.csv')
```

```python
df.summary()
```

```python
df.col_names
```

```python
x = df.col_names
x.remove('net')
x.remove('target')
y='target'
```

```python
df['target'] = df['target'].asfactor()
df['release_month'] = df['release_month'].asfactor()
df['release_year'] = df['release_year'].asfactor()
df['release_dow'] = df['release_dow'].asfactor()
```

```python
train, valid, test = df.split_frame(ratios=[0.7, 0.1])
print(df.shape)
print(train.shape)
print(valid.shape)
print(test.shape)
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
drf = H2ORandomForestEstimator()
drf.train(x=x, y=y, training_frame=train, validation_frame=valid)
```

```python
drf.model_performance(test_data=test)
```

```python
drf.score_history()
```

```python
drf.predict(test)
```
